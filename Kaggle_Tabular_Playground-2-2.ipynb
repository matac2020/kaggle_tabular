```python
#There are lots of features. Probably many of them doesn't have valuable information for a model. Let's decrease it
```

```python
#Feature Selector didn't worked as above. We manually try to decrease the number of features
```

```python
#Variance threshold
#We can elimate features by looking at their variance threshold. Example below eliminates features with 
#only 1 unique value
```

```python
x_train_scale
x_test_scale
y_train
y_test
```

```python
from sklearn.feature_selection import VarianceThreshold

selector = VarianceThreshold()
print("Original feature shape:", x_train_scale.shape)
select = selector.fit(x_train_scale)
x_train_new = select.transform(x_train_scale)
x_test_new = select.transform(x_test_scale)

print("Transformed feature shape:", x_train_new.shape)
```

```python
#Univariate feature selection
# Next we are going to use f_classif test
#The f_classif test is based on the Analysis of Variance (ANOVA) statistical test which compares the means of a 
#number of groups, or in our case features, and determines whether any of those means are statistically
#significant from one another. (the default number of feature is 10)
```

```python
from sklearn.feature_selection import SelectKBest

selector = SelectKBest(k=15)

print("Original feature shape:", x_train_new.shape)

select = selector.fit(x_train_new,y_train)
x_train_new2 = select.transform(x_train_new)
x_test_new2 = select.transform(x_test_new)

print("Transformed feature shape:", x_train_new2.shape)
```

```python
#Logistic Regression
lm = LogisticRegression()
lm.fit(x_train_new2, y_train)
y_pred = lm.predict(x_test_new2)
print('Test Set Weighted Accuracy:',classification_report(y_test, y_pred, output_dict=True)['weighted avg']["precision"])
print(" ")
print(classification_report(y_test, y_pred))
```

```python
#Decision Tree#
clf = DecisionTreeClassifier()
clf = clf.fit(x_train_new2,y_train)
y_pred = clf.predict(x_test_new2)
print('Test Set Weighted Accuracy:',classification_report(y_test, y_pred, output_dict=True)['weighted avg']["precision"])
print(" ")
print(classification_report(y_test, y_pred))
```

```python
#Random Forest
rf = RandomForestClassifier()
rf.fit(x_train_new2, y_train)
y_pred = rf.predict(x_test_new2)
print('Test Set Weighted Accuracy:',classification_report(y_test, y_pred, output_dict=True)['weighted avg']["precision"])
print(" ")
print(classification_report(y_test, y_pred))
```

```python
#AdaBoost#
AdaBoost = AdaBoostClassifier()
AdaBoost.fit(x_train_new2,y_train)
y_pred=AdaBoost.predict(x_test_new2)
print('Test Set Weighted Accuracy:',classification_report(y_test, y_pred, output_dict=True)['weighted avg']["precision"])
print(" ")
print(classification_report(y_test, y_pred))
```
